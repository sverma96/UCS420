Q2) Create a dataset of five subjects and scores.

a) Convert it into a Pandas DataFrame.

```python
import pandas as pd
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Geography'],
    'Score': [85, 92, 78, 88, 91]
}

# Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(data)

print(df)
```

b) Plot the scores using a Seaborn bar plot with:

*   Different colors for each bar.
*   Annotations on top of each bar.


*   Title, axis labels, and grid.





```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the dataset of subjects and scores
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Geography'],
    'Score': [85, 92, 78, 88, 91]
}

# Set the Seaborn style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')

# Annotate each bar with the score value
for i in range(len(df)):
    bar_plot.text(i, df['Score'][i] + 1, df['Score'][i], ha='center', color='black')

# Add title, axis labels, and grid
plt.title("Subject Scores", fontsize=16)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
