Q1) Ask the user to enter a value (e.g., M) for a mathematical function.

a) Generate x values from -10 to 10 using np.linspace().

```python
import numpy as np
M = float(input("Enter a value for M: "))
x_values = np.linspace(-10, 10, 100)
print("Generated x values:", x_values)
```

b) Compute y values for:


*   y= M⋅x
2
*   y=M⋅sin(x)



```python
M = float(input("Enter a value for M: "))
# Compute y values for y = M * x^2
y1_values = M * x_values**2
# Compute y values for y = M * sin(x)
y2_values = M * np.sin(x_values)

print("y = M * x^2 values:", y1_values )
print("\n y = M * sin(x) values:", y2_values)
```

c) Plot both functions in a single figure:

*   Use different colors and line styles.
*   Add legend, grid, and title.



```python
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))


# Plot both functions in a single figure
plt.figure(figsize=(8, 6))

# Plot y = M * x^2 with blue color and dashed line style
plt.plot(x_values, y1_values, label=r'$y = M \cdot x^2$', color='blue', linestyle='--')

# Plot y = M * sin(x) with orange color and solid line style
plt.plot(x_values, y2_values, label=r'$y = M \cdot \sin(x)$', color='orange', linestyle='-')

# Add title
plt.title(f"Plots of $y = M \cdot x^2$ and $y = M \cdot \sin(x)$ for M = {M}")

# Add x and y axis labels
plt.xlabel('x')
plt.ylabel('y')

# Show grid
plt.grid(True)

# Add legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
