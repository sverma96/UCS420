Q3) a) Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

```python
import numpy as np
import matplotlib.pyplot as plt

# Set your roll number as the seed for NumPy (replace 12345 with your actual roll number)
roll_number = 12345  # Replace this with your roll number
np.random.seed(roll_number)

# Generate a dataset of 50 random values using np.random.randn()
data = np.random.randn(50)
print(data)
```

b) Create a 2x2 subplot layout displaying:

*   Line plot showing cumulative sum.
*   Scatter plot with random noise.

*   Customize with titles, axis labels, and grids.






```python
# Create a 2x2 subplot layout
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# 1st plot: Line plot showing cumulative sum
axs[0].plot(np.cumsum(data), label='Cumulative Sum', color='blue')
axs[0].set_title('Cumulative Sum')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Sum')
axs[0].grid(True)

# 2nd plot: Scatter plot with random noise
axs[1].scatter(range(50), data, label='Random Noise', color='orange')
axs[1].set_title('Scatter Plot with Random Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Value')
axs[1].grid(True)

# Add legends
axs[0].legend()
axs[1].legend()

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```
