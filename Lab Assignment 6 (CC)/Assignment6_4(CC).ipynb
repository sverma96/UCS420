*   Apply ‘seaborn’ library to do the following.


1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.







```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="b", ax=axes[0])
axes[0].set_title("Total Profit per Month")
axes[0].set_xlabel("Month Number")
axes[0].set_ylabel("Total Profit")
axes[0].grid()

for column in df.columns[1:-1]:
    sns.lineplot(x=df["month_number"], y=df[column], label=column, ax=axes[1])
axes[1].set_title("Product Sales per Month")
axes[1].set_xlabel("Month Number")
axes[1].set_ylabel("Sales")
axes[1].legend()
axes[1].grid()

df.set_index("month_number").plot(kind="bar", ax=axes[2], colormap="viridis")
axes[2].set_title("Sales Data")
axes[2].set_xlabel("Month Number")
axes[2].set_ylabel("Values")
axes[2].grid(axis='y')

plt.tight_layout()
plt.show()
```
