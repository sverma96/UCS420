```python
# ANS 6.
import pandas as pd
employeeData={
    'Employee_ID':[101,102,103,104,105],
    'Name':["Alice","Bob","Charlie","Diana","Edward"],
    'Department':["HR","IT","IT","Marketing","Sales"],
    'Age':[29,34,41,28,38],
    'Salary':[50000,70000,65000,55000,60000],
    'Years_of_Experience':[4,8,10,3,12],
    'Joining_Date':["2020-03-15","2017-07-19","2013-06-01","2021-02-10"," 2010-11-25"],
    'Gender':["Female","Male","Male","Female","Male"],
    'Bonus':[5000,7000,6000,4500,5000],
    'Rating':[4.5,4.0,3.8,4.7,3.5]
}
emp=pd.DataFrame(employeeData)
pd.set_option('display.max_columns', None)   # for all columns to be aligned
print(f"The dataset is follwoing: \n{emp}")
r,c=emp.shape
print()
print(f"(a) Shape of DataFrame i.e. no. of rows are :{r} and no. of columns are: {c}\n")

print(f"(b) Summary of the DataFrame that includes the data types and non-null counts for each column:\n")
print(emp.info(),"\n")

print(f"(c) Descriptive statistics:\n{emp.describe()}\n")

print(f"(d) First 5 rows:\n{emp.head(5)} \nand last 3 rows:\n{emp.tail(3)}\n")

print(f"(e) Calculating the following:\n   (i) Average salary: {emp['Salary'].mean()}\n   (ii) Total bonus paid: {emp['Bonus'].sum()}\n   (iii) The youngest employee is {emp['Age'].min()} years old\n   (iv) The highest performance rating is {emp['Rating'].max()}\n")

print(f"(f) Sorting DataFrame by Salary in descending order: \n{emp.sort_values(by='Salary',ascending =False)}\n")

def categ(rating):
    if (rating>=4.5):
        return "Excellent"
    elif (rating>=4.0):
        return "Good"
    else:
        return "Average"
emp["Performance"]=emp['Rating'].apply(categ)
print(f"(g) Adding new column for categorizing employees:\n{emp}\n")

print(f"(h) Missing values are:{emp.isnull().sum()}\n")

emp.rename(columns={'Employee_ID':'ID'}, inplace=True)
print(f"(i) Renaming Employee_ID column: \n{emp}\n")

print(f"(j) Finding employees with: \n   (i) More than 5 years of experience: \n{emp[emp["Years_of_Experience"]>5]}\n   (ii) Belonging to IT department: \n{emp[emp["Department"]=="IT"]}\n")

emp["Tax"]=emp["Salary"]*0.1
print(f"(k) Adding a Tax column: \n{emp}\n")

emp.to_csv("employees.csv",index=False)
print(f"(k) Modified Employee data saved to file named \"employees.csv\"")
```
